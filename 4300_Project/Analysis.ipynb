I wanna compare my own liked songs to the 'average' song and see how they are different, so i found a dataset of about 136000 songs and its audio features. I will put this in mongo just in case i need to reimport it into a differrent notebook

```python
#importing a dataset of a lot of songs for comparison
lot_of_songs = pd.read_csv("SpotifyAudioFeaturesApril2019.csv")
lot_of_songs
```

```python
import pymongo
from pymongo import MongoClient
client = MongoClient()
spotify_db = client.spotify
collection_mass = spotify_db.mass_collection
for row in lot_of_songs.to_dict(orient="row"):
    collection_mass.insert_one(row)
```

```python
#find means and stds of each feature in all relevant dataframes
#relevant dataframes
my_library.head()
```

```python
relevant_cols = list(my_library.columns)
relevant_cols.remove("Target")
my_library_means = dict(zip(relevant_cols,[my_library[i].mean() for i in relevant_cols]))
my_library_means
```

```python
disliked_songs.head()
```

```python
disliked_songs_means = dict(zip(relevant_cols,[disliked_songs[i].mean() for i in relevant_cols]))
disliked_songs_means
```

```python
#accessing mongo for the others
good_songs_collection=spotify_db.good_songs
hits = pd.DataFrame(list(good_songs_collection.find()))
hits.head()
```

```python
hits_means = dict(zip(relevant_cols,[hits[i].mean() for i in relevant_cols]))
hits_means
```

```python
#accessing mongo again
good_songs_collection=spotify_db.bad_songs
flops = pd.DataFrame(list(bad_songs_collection.find()))
flops.head()
```

```python
flops_means = dict(zip(relevant_cols,[flops[i].mean() for i in relevant_cols]))
flops_means
```

```python
lot_of_means = dict(zip(relevant_cols,[lot_of_songs[i].mean() for i in relevant_cols]))
lot_of_means
```

```python
#finding std of the larger song database to find z scores of everything and make a visualization this is the
#standard for the rest of the music
lot_of_stds = dict(zip(relevant_cols,[lot_of_songs[i].std() for i in relevant_cols]))
lot_of_stds
```

```python
my_library_z_scores = {}
for k in my_library_means.keys():
    my_library_z_scores[k] = (my_library_means[k] - lot_of_means[k])/lot_of_stds[k]
my_library_z_scores
```

```python
disliked_z_scores = {}
for k in my_library_means.keys():
    disliked_z_scores[k] = (disliked_songs_means[k] - lot_of_means[k])/lot_of_stds[k]
disliked_z_scores
```

```python
hit_z_scores = {}
for k in my_library_means.keys():
    hit_z_scores[k] = (hits_means[k] - lot_of_means[k])/lot_of_stds[k]
hit_z_scores
```

```python
flop_z_scores = {}
for k in my_library_means.keys():
    flop_z_scores[k] = (flops_means[k] - lot_of_means[k])/lot_of_stds[k]
flop_z_scores
```

```python
%matplotlib
import matplotlib.pyplot as plt
```

```python
#plot will open in another window
plt.clf()

xlabels = list(my_library_z_scores.keys())
indices = range(len(xlabels))
width = np.min(np.diff(indices))/4.

p1 = plt.bar(indices-width/1., my_library_z_scores.values(), width = width)
p2 = plt.bar(indices+width/1., flop_z_scores.values(), width = width)
p3 = plt.bar(indices, hit_z_scores.values(), width = width)
locs, labels=plt.xticks()
plt.xticks(np.arange(13), xlabels)
plt.legend((p1[0], p2[0], p3[0]), ('Liked', 'Flops', 'Hits'))
plt.show()
```

Plot shows features of my_library, hits, and flops compared to normal songs, for easy comparison. Now, after running the ML model in the mongoML notebook, lets compare this to the actual songs recommended to me

```python
#retrieve mongo collection created in mongoML notebook
final_recs_collection=spotify_db.final_recs
final_recs = pd.DataFrame(list(final_recs_collection.find()))
final_recs.head()
```

```python
finalrecs_means = dict(zip(relevant_cols,[final_recs[i].mean() for i in relevant_cols]))
```

```python
finalrecs_z_scores = {}
for k in my_library_means.keys():
    finalrecs_z_scores[k] = (finalrecs_means[k] - lot_of_means[k])/lot_of_stds[k]
finalrecs_z_scores
```

```python
#plot will open in another window
plt.clf()

xlabels = list(my_library_z_scores.keys())
indices = range(len(xlabels))
width = np.min(np.diff(indices))/3.

p1 = plt.bar(indices-width/2., my_library_z_scores.values(), width = width)
p2 = plt.bar(indices+width/2., finalrecs_z_scores.values(), width = width)
locs, labels=plt.xticks()
plt.xticks(np.arange(13), xlabels)
plt.legend((p1[0], p2[0]), ('Liked', 'Recommendations'))
plt.show()
```
