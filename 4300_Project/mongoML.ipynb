```python
import pymongo
from pymongo import MongoClient
client = MongoClient()
spotify_db = client.spotify
library_collection = spotify_db.library
recommendation_collection = spotify_db.recs
bad_songs_collection = spotify_db.bad_songs
import pandas as pd
import numpy as np
import pprint
```

```python
#import into dataframe from mongo
my_library = pd.DataFrame(list(library_collection.find()))
my_library
```

```python
#setting target for ML and removing irrelevant data
my_library["Target"] = 1
my_library = my_library.drop(columns=["_id", "id", "type", "track_href", "analysis_url"])
my_library = my_library.drop(columns=["uri"])
my_library
```

```python
#doing the same for 'disliked' songs, the first line imports the data from mongo
disliked_songs = pd.DataFrame(list(bad_songs_collection.find()))
disliked_songs["Target"] = 0
disliked_songs = disliked_songs.drop(columns=["_id"])
disliked_songs
```

```python
#appending the dataframes into 1
all_test_data = my_library.append(disliked_songs)
all_test_data = all_test_data.drop(columns=["Track"])
#shuffles the dataframe
all_test_data = all_test_data.sample(frac=1).reset_index(drop=True)
all_test_data
```

```python
#time to build a classifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

estimators = {"k-Nearest Neighbor": KNeighborsClassifier(), 
              "Support Vector Machine" : SVC(gamma=0.002), 
              "Gaussian Naive Bayes" : GaussianNB(), 
              "Decision Tree" : DecisionTreeClassifier()}
```

```python
all_test_data = all_test_data.dropna()
```

```python
features = all_test_data.iloc[0:, 0:13]
features
```

```python
target = all_test_data["Target"]
target
```

```python
from sklearn.model_selection import train_test_split
#splits data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=3000)
```

```python


def classifiers_percentage_split():
    #iterates over estimators
    for m in estimators.keys():
        #fits model with data
        model = estimators[m].fit(X=X_train, y=y_train)
        #predicts accuracy
        accuracy = model.score(X_test, y_test)
        #print statement and formatting
        print(m + ":\n         Prediction accuracy on the test data:", f"{accuracy:.2%}")
```

```python
classifiers_percentage_split()
```

So far the decision tree has the highest accuracy, will probably end up using that

```python
#cross validation
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

def classifiers_cross_validation():
    kfold = KFold(n_splits=10, random_state=3000, shuffle=True)
    #iterates over estimators
    for m in estimators.keys():
        #fits model with data
        model = estimators[m].fit(X=X_train, y=y_train)
        scores = cross_val_score(estimator=model, X=features, y=target, cv=kfold)
        #print statement and formatting
        print(m + ":\n         " + f'Mean accuracy: {scores.mean():.2%}' + f'standard deviation={scores.std():.2%}')
```

```python
classifiers_cross_validation()
```

most data is already normalized as avaialble through the spotify API, but we will normalize it all again and then test
the model

```python
from sklearn import preprocessing
def preprocess():
    scaler = preprocessing.MinMaxScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.fit_transform(X_test)
    #iterates over estimators
    for m in estimators.keys():
        #fits model with data
        model = estimators[m].fit(X=X_train_scaled, y=y_train)
        #predicts accuracy
        accuracy = model.score(X_test_scaled, y_test)
        
        
        #print statement and formatting
        print(m)
        print(accuracy)
        
    return X_train_scaled, X_test_scaled
X_train_scaled, X_test_scaled = preprocess()
```

```python
#run cross validation with scaled data

def classifiers_cross_validation2():
    kfold = KFold(n_splits=10, random_state=3000, shuffle=True)
    #iterates over estimators
    for m in estimators.keys():
        #fits model with data
        model = estimators[m].fit(X=X_train_scaled, y=y_train)
        scores = cross_val_score(estimator=model, X=features, y=target, cv=kfold)
        #print statement and formatting
        print(m + ":\n         " + f'Mean accuracy: {scores.mean():.2%}' + f'standard deviation={scores.std():.2%}')
classifiers_cross_validation2()
```

```python
#After cross-validation again, decision tree still seems better, we will use that
```

```python
#rapid feature elimination
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeRegressor
#below we return the 9 most important features but we will use all 12 in the actual model
def RFE_feature_selection():
    select = RFE(DecisionTreeRegressor(random_state = 3000), n_features_to_select = 3)
    #fit the RFE selector to the training data
    select.fit(X_train, y_train)
    #transform training and testing sets so only the selected features are retained
    X_train_selected = select.transform(X_train_scaled)
    X_test_selected = select.transform(X_test_scaled)

    
    #fits model with data
    model = DecisionTreeClassifier().fit(X=X_train_selected, y=y_train)
    #predicts accuracy
    accuracy = model.score(X_test_selected, y_test)


    print("Selected features after RFE:")
    for i in range(0, len(select.get_support())):
        if select.get_support()[i] == True:
            print("\t" + str(features.columns[i]))

    print("Accuracy with selected features:")
    print(accuracy)

RFE_feature_selection()
```

Danceability seems to be an important influence on whether or not I'll like a song, I think it makes sense

```python

#tuning decision tree
from sklearn.model_selection import GridSearchCV
param_grid = {"max_depth":[1, 10, 100], "min_samples_split": [2, 10, 100]}

def grid_search_dectree():
    X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=3000)
    grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5)
    
    #fits the grid search object on the training data 
    grid_search.fit(X=X_train, y=y_train)
    
    #printing
    print("Best parameters: ", grid_search.best_params_)
    print("Training set score with best parameters: ", grid_search.score(X_train, y_train))
    print("Test set score with best parameters: ", grid_search.score(X_test, y_test))
    print("Best cross-validation score: ", grid_search.best_score_)
    
grid_search_dectree()
```

```python
#now i will prepare the dataframe of potential recommendations, so that the classfier can run on all songs here
#and recommend me some of them
#the first line of code imports it from Mongo
potential_recommendations = pd.DataFrame(list(recommendation_collection.find()))
potential_recommendations = potential_recommendations.drop(columns=["_id", 'type', 'id', 'uri', 'track_href'])
potential_recommendations = potential_recommendations.drop(columns=["analysis_url"])
potential_recommendations.index = potential_recommendations["Track"]
potential_recommendations
```

```python
#Final function to predict which songs i'll like
#model used
model = DecisionTreeClassifier().fit(X=X_train_scaled, y=y_train)

#List of all features
feat_list = list(features.columns)

def will_like(song):
    tup1 = []
    #iterate through features
    for feat in feat_list:
        #add each feature to each list
        tup1.append(potential_recommendations[feat][song])
    result1 = model.predict([tup1])[0]
    #returns whether or not the model recommends the song
    if result1 == 1:
        return "You may like this song"
    else:
        return "Not recommended"
```

```python
%matplotlib
import matplotlib.pyplot as plt
```

```python
#plots feature importance
#plot will open in new window
labels = features.columns
sizes = model.feature_importances_

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%')
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
#plots ROC curve for model results
import sklearn.metrics as metrics
# calculate the fpr and tpr for all thresholds of the classification
probs = model.predict_proba(X_test_scaled)
preds = probs[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, preds)
roc_auc = metrics.auc(fpr, tpr)

# method I: plt
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
#removes songs from potential recommendtions if already in library
reclist = list(potential_recommendations["Track"])
for i in list(my_library['Track']):
    if i in reclist:
        
        reclist.remove(i)
```

```python
#runs function on every song and returns the songs you may like
final_recommendations = []
for song in reclist:
    if will_like(song) == "You may like this song":
        final_recommendations.append(song)
final_recommendations
```
